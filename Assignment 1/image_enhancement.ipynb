```python
image = cv2.imread("./data/side.jpg")


def enhance(img, skin=True, sky=True, shadow=True):
    if skin:
        img = correct_face(img)
    if sky:
        img = sky_enhancement(img)
    if shadow:
        img = shadow_enhance(img)
    return img


img_out = enhance(image, sky=False, shadow=False)
cv2.imshow("Original", image)
cv2.imshow("Final", img_out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
